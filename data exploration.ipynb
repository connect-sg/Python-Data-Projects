# You are given a dataset, which is present in the LMS, containing the number of hurricanes occurring in the United States along the coast of the Atlantic. Load the data from the dataset into your program and plot a Bar Graph of the data, taking the Year as the x-axis and the number of hurricanes occurring as the Y-axis.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('dataset/Hurricanes.csv')
```

```python
data
```

```python
plt.bar(data['Year'],data['Hurricanes'])
plt.ylabel('no of hurricanes')
plt.xlabel('year')
```

# Plot a pie-chart of the number of models released by every manufacturer, recorded in the data provide. Also mention the name of the manufacture with the largest releases.

```python
cars = pd.read_csv('dataset/Cars2015.csv')
```

```python
cars
```

```python
car_by_make = cars['Model'].groupby(cars['Make']).count()
```

```python
car_by_make.plot.pie(figsize =(10,10))
```
