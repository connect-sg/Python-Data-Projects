```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = [1,2,3,4]
y = [20, 21, 20.5, 20.8]
plt.scatter(X,y)
```

```python
plt.plot(X,y)
```

```python
plt.plot(X,y,linestyle="dashed", marker="*", color="red")
```

```python
plt.plot(X,y,linestyle="dashed", marker="*", color="red")
plt.xlabel('X')
plt.ylabel('y')
plt.xlim(0,5)
plt.xticks([1,2,3,4])
plt.ylim(20,21)
plt.yticks([20.0,20.3,20.6,20.9,21])
plt.title("this is x vs y plot")
plt.show()
```

```python
fig= plt.figure(figsize=(4,5),dpi= 100)
plt.rc('font',size = 14)
y_error = [0.12, 0.13, 0.2, 0.1]
plt.errorbar(X, y, yerr=y_error, linestyle="None", marker="None", color="green")
plt.plot(X,y,linestyle="dashed", marker="*", color="red")
plt.xlabel('X')
plt.ylabel('y')
plt.xlim(0,5)
plt.xticks([1,2,3,4])
plt.ylim(20,21)
plt.yticks([20.0,20.3,20.6,20.9,21])
plt.title("this is x vs y plot")
plt.show()
```
